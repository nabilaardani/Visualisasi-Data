## 1. Import dan tampilkan dataset dengan judul auto.csv. Buatlah scatter plot biasa menggunakan variabel mpg dengan weight dan bagaimana interpretasinya

```python
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
auto = pd.read_csv("/content/auto.csv")
print(auto)
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(auto["weight"], auto["mpg"], color='blue')
plt.title("MPG vs Weight", fontsize=25)
plt.xlabel("Weight", fontsize=15)
plt.ylabel("MPG", fontsize=15)
plt.xlim([0, 5000])
plt.ylim([0, 40])
plt.grid()
plt.show()
```

### 2. Buatlah dan interpretasikan box plot pada dataset tips seperti contoh antara variabel total_bill dengan day, dan tip dengan time.


```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.catplot(data=tips, x="total_bill", y="day", kind="box")
```

```python
sns.catplot(data=tips, x="tip", y="time", kind="box")
```

## 3. Buatlah dan interpretasikan gabungan scatter plot pada dataset tips seperti contoh antara variabel total_bill, size dengan sex dan day 

```python
sns.relplot(
    data=tips,
    x="total_bill", y="size", hue="sex", col="day",)
```
